```python
#Vinay Bhargav Jampa
#Analysis

import pandas as pd


file_path=("Data Analyst Intern Assignment - Excel.xlsx")

user_data=pd.read_excel(file_path, sheet_name="UserDetails.csv")
cooking_data=pd.read_excel(file_path,sheet_name="CookingSessions.csv")
order_data=pd.read_excel(file_path,sheet_name="OrderDetails.csv")

print(user_data.info())
print(cooking_data.info())
print(order_data.info())
```

```python
# Display the first few rows to understand the structure
print(user_data.head())
print(cooking_data.head())
print(order_data.head())

# Check for missing values
print(user_data.isnull().sum())
print(cooking_data.isnull().sum())
print(order_data.isnull().sum())
```

```python
# Convert columns to appropriate types (e.g., dates, categorical variables)
user_data['user_id'] = user_data['User ID'].astype(str)
cooking_data['session_id'] = cooking_data['Session ID'].astype(str)
order_data['order_id'] = order_data['Order ID'].astype(str)


order_data['Rating'].fillna(order_data['Rating'].mean(), inplace=True)
```

```python
# Merge datasets
merged_data = pd.merge(user_data, cooking_data, on='User ID', how='left')
merged_data = pd.merge(merged_data, order_data, on='User ID', how='left')
print(merged_data.info())
merged_data
```

```python
summary=merged_data.isnull().sum()
summary
```

```python
merged_data = merged_data.dropna(subset=[
    'Session ID_x', 'Dish Name_x', 'Meal Type_x', 'Session Start', 'Session End',
    'Duration (mins)', 'Session Rating', 'session_id', 'Order ID', 'Order Date', 
    'Meal Type_y', 'Dish Name_y', 'Order Status', 'Amount (USD)', 'Time of Day',
    'Rating', 'Session ID_y', 'order_id'
])

print(merged_data.isnull().sum())
```

```python
merged_data = merged_data.drop(columns=[
    'Session ID_x', 'Dish Name_x', 'Meal Type_x', 'Session ID_y', 'Dish Name_y', 'Meal Type_y'
])

# You might want to keep one version of the session and order ID columns, for example:
# If the session and order ID columns are similar, you can choose which one to keep
# Keep 'session_id' and 'order_id' columns from the merge.
merged_data = merged_data.rename(columns={'session_id': 'session_id_final', 'order_id': 'order_id_final'})

# Verify the cleaned dataset
print(merged_data.head())
```

```python
# Summary statistics of numerical columns
print(merged_data.describe())

```

```python
# Summary statistics for categorical columns
print(merged_data['Favorite Meal'].value_counts())
print(merged_data['Order Status'].value_counts())
```

```python
#Exploratory Data Analysis (EDA)

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Age distribution
sns.histplot(merged_data['Age'], kde=True)
plt.title('Age Distribution of Users')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

# Meal preference vs. total orders
meal_order_counts = merged_data.groupby('Favorite Meal')['Total Orders'].sum().sort_values(ascending=False)
meal_order_counts.plot(kind='bar')
plt.title('Total Orders by Favorite Meal')
plt.ylabel('Total Orders')
plt.show()

# Age vs. total orders
sns.scatterplot(x='Age', y='Total Orders', data=merged_data)
plt.title('Age vs. Total Orders')
plt.xlabel('Age')
plt.ylabel('Total Orders')
plt.show()
```

```python
# Session duration vs. order amount
sns.scatterplot(x='Duration (mins)', y='Amount (USD)', data=merged_data)
plt.title('Session Duration vs. Order Amount')
plt.xlabel('Session Duration (mins)')
plt.ylabel('Order Amount (USD)')
plt.show()
```

```python
# Time of day vs. total order amount
time_of_day_order = merged_data.groupby('Time of Day')['Amount (USD)'].sum().sort_values(ascending=False)
time_of_day_order.plot(kind='bar')
plt.title('Total Order Amount by Time of Day')
plt.ylabel('Total Order Amount (USD)')
plt.show()
```
